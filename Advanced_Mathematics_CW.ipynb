# **Question** **01** - Application of Differentiation in AI Model Optimization

(a).

```python
import sympy as sp   # Import the symbolic mathematics library

x, y = sp.symbols('x y')   # Define x and y as symbolic variables
L = sp.exp(x**2 + y**2) + sp.cos(x * y)   # Define the loss function L(x, y)
partial_x = sp.diff(L, x)       # Compute the partial derivative of L with respect to x
partial_y = sp.diff(L, y)   # Compute the partial derivative of L with respect to y

print("Partial Derivative of L with respect to x (∂L/∂x):")
sp.pprint(partial_x)  # Formatted output

print("\nPartial Derivative of L with respect to y (∂L/∂y):")
sp.pprint(partial_y)
```

```python
import numpy as np

def compute_gradient(x, y):
    """Compute the gradient vector of L(x, y) = e^(x^2 + y^2) + cos(xy)."""
    exp_term = np.exp(x**2 + y**2)
    sin_term = np.sin(x * y)

    dL_dx = 2 * x * exp_term - y * sin_term
    dL_dy = 2 * y * exp_term - x * sin_term

    return np.array([dL_dx, dL_dy])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def A(t):
    return 2 * t**2 * np.exp(-t)

# Create the t values for plotting
t_values = np.linspace(0, 4, 100)
A_values = A(t_values)

# Calculate the area under the curve using the trapezoidal rule
area = np.trapz(A_values, t_values)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(t_values, A_values, label='$A(t) = 2t^2 e^{-t}$')

# Fill the area under the curve
plt.fill_between(t_values, A_values, where=[(t >= 0) and (t <= 4) for t in t_values], color='yellow', alpha=0.5)

# Annotate the graph
plt.text(2, 1.3, f'Area under the curve = {area:.2f}', horizontalalignment='center', fontsize=12)

# Labels and title
plt.xlabel('t')
plt.ylabel('A(t)')
plt.title('Plot of $A(t) = 2t^2 e^{-t}$ with Area Under the Curve')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Output the area
print(f'Area under the curve: {area:.2f}')
```

# **Question** **03** - Series Convergence and Numerical Approximation

(a). and (b).

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the series approximation of ln(1+x)
def ln_approximation(x, terms=10):
    result = 0
    for n in range(1, terms + 1):
        result += ((-1)**(n+1)) * (x**n) / n
    return result

# Define the x values in the range [-1, 1]
x_values = np.linspace(-1, 1, 400)

# Compute the series approximation for each x
approx_values = ln_approximation(x_values)

# Compute the actual ln(1+x) values
actual_values = np.log(1 + x_values)

# Plot the approximation and actual values
plt.figure(figsize=(8, 6))
plt.plot(x_values, approx_values, label="Series Approximation", color='blue', linestyle='dashed')
plt.plot(x_values, actual_values, label="Actual ln(1+x)", color='red')
plt.title("Series Approximation vs Actual ln(1+x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

(c).

The series for ln(1+x) behaves differently for different values of x:

1. Convergence of the Series:
   - The series for ln(1+x) works well for small values of |x|, meaning when x is close to 0.
   - As |x| gets larger (especially near 1), the error between the series approximation and the actual value increases.
   - For values of |x| > 1, the series does not work well and the error becomes very large.

2. Error Analysis:
   - We calculate the difference between the series approximation and the actual value of ln(1+x) to see how the error changes with x.
   - This error can be plotted to show how it increases for larger values of |x|.

3. Error Behavior:
   - For small |x| (near 0), the series approximation is very accurate, and the error is small. This is because higher-order terms in the series have little effect near 0.
   - As |x| approaches 1 (either positive or negative), the error grows because the series takes more terms to match the actual value.
   - When |x| > 1, the series starts to diverge, and the error increases quickly.


The series approximates ln(1+x) well for small values of |x|. As |x| increases, the approximation gets worse, and for |x| ≥ 1, the series is no longer accurate.


```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the series approximation of ln(1+x)
def ln_approximation(x, terms=10):
    result = 0
    for n in range(1, terms + 1):
        result += ((-1)**(n+1)) * (x**n) / n
    return result

# Define the x values in the range [-1, 1]
x_values = np.linspace(-1, 1, 400)

# Compute the series approximation for each x
approx_values = ln_approximation(x_values)

# Compute the actual ln(1+x) values
actual_values = np.log(1 + x_values)

# Calculate the absolute error
error_values = np.abs(actual_values - approx_values)

# Plot the error between the approximation and the actual values
plt.figure(figsize=(8, 6))
plt.plot(x_values, error_values, label="Error between Approximation and Actual", color='green')
plt.title("Error Analysis of the Series Approximation for ln(1+x)")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.legend()
plt.grid(True)
plt.show()
```

# **Question** **04** -  Fourier Transform for Data Filtering

(a).

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal parameters
t = np.linspace(0, 1, 500, endpoint=False)  # Time vector
signal = np.sin(2 * np.pi * t) + 0.5 * np.sin(10 * np.pi * t) + 0.3 * np.sin(50 * np.pi * t)

# (a) Perform Fourier Transform and plot frequency spectrum
fft_signal = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(t), d=(t[1] - t[0]))  # Frequency vector
magnitude = np.abs(fft_signal)  # Magnitude of FFT

# Plot the frequency spectrum
plt.figure(figsize=(10, 6))
plt.plot(frequencies[:len(frequencies)//2], magnitude[:len(magnitude)//2])  # Positive frequencies
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```

(b).

```python
# (b) Apply low-pass filter to remove frequencies above 20 Hz
low_pass_filter = np.abs(frequencies) <= 20  # Mask for frequencies <= 20 Hz
filtered_fft_signal = fft_signal * low_pass_filter
filtered_signal = np.fft.ifft(filtered_fft_signal)  # Reconstruct signal

# Plot the original and filtered signals
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label="Original Signal")
plt.plot(t, filtered_signal.real, label="Filtered Signal", linestyle='--', color="red")
plt.title("Original vs Filtered Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

(c).

Filtering impacts the signal by removing unwanted frequency components, making the signal cleaner and easier to interpret. For example, applying a low-pass filter in this case eliminates high-frequency noise, retaining the essential lower-frequency components of the signal.

In real-world scenarios, filtering is crucial in data science applications such as:
1. **Audio Processing**: Reducing noise in audio recordings, such as removing background sounds in voice communication systems.
2. **Image Processing**: Smoothing images by eliminating high-frequency noise, improving visual quality.
3. **Biomedical Signal Analysis**: Cleaning signals like EEG or ECG by removing unwanted noise while preserving critical diagnostic information.
4. **Telecommunications**: Enhancing signal quality by filtering out interference and ensuring reliable communication.

Overall, filtering enhances the accuracy and usability of data by focusing on relevant information while discarding irrelevant or noisy components.


# **Question** **05** - Image Processing Using Fourier and Discrete Cosine Transforms

(a).

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread('/content/drive/MyDrive/Math CW/Image (3).png', cv2.IMREAD_GRAYSCALE)

# Compute the 2D FFT and shift the zero frequency component to the center
f = np.fft.fft2(image)
fshift = np.fft.fftshift(f)

# Create a high-pass filter: mask with center blocked
rows, cols = image.shape
crow, ccol = rows // 2, cols // 2
mask = np.ones((rows, cols), np.uint8)
r = 30  # Radius for the blocked center
mask[crow - r:crow + r, ccol - r:ccol + r] = 0

# Apply the mask and inverse shift
fshift_filtered = fshift * mask
f_ishift = np.fft.ifftshift(fshift_filtered)

# Perform the Inverse FFT to get the edge-detected image
edge_image = np.abs(np.fft.ifft2(f_ishift))

# Visualize the edge-detected image
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(edge_image, cmap='gray'), plt.title('Edge Detected Image')
plt.show()
```

(b).

```python
def gaussian_low_pass_filter(shape, sigma):
    rows, cols = shape
    crow, ccol = rows // 2, cols // 2
    x, y = np.meshgrid(np.linspace(-ccol, ccol, cols), np.linspace(-crow, crow, rows))
    gaussian_filter = np.exp(-(x**2 + y**2) / (2 * sigma**2))
    return gaussian_filter

# Gaussian blur in frequency domain
sigma = 20  # Standard deviation for Gaussian filter
gaussian_filter = gaussian_low_pass_filter(image.shape, sigma)

# Apply Gaussian filter in the frequency domain
fshift_blur = fshift * gaussian_filter
f_ishift_blur = np.fft.ifftshift(fshift_blur)
blurred_image_fft = np.abs(np.fft.ifft2(f_ishift_blur))

# Standard Gaussian blur using OpenCV
blurred_image_cv = cv2.GaussianBlur(image, (15, 15), sigma)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.subplot(131), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(132), plt.imshow(blurred_image_fft, cmap='gray'), plt.title('Gaussian Blur ( Fourier)')
plt.subplot(133), plt.imshow(blurred_image_cv, cmap='gray'), plt.title('Gaussian Blur (Standard)')
plt.show()
```

```python
from scipy.fftpack import dct, idct

# Apply DCT to the image
dct_transformed = dct(dct(image.T, norm='ortho').T, norm='ortho')

# Scale down the image in DCT domain
scaled_dct = np.zeros_like(dct_transformed)
scaled_dct[:240, :240] = dct_transformed[:240, :240]

# Inverse DCT to reconstruct the scaled image
reconstructed_image = idct(idct(scaled_dct.T, norm='ortho').T, norm='ortho')

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.subplot(1, 3, 2)
plt.title("DCT Scaled Image (240x240)")
plt.imshow(reconstructed_image, cmap="gray")
plt.subplot(1, 3, 3)
plt.title("Difference")
plt.imshow(np.abs(image - reconstructed_image), cmap="gray")
plt.show()
```

```python

from scipy.fftpack import dct, idct

# 1. Global DCT Compression: Simulating Ringing Artifacts
def quantize_dct(dct_coeffs, threshold=50):

    quantized = dct_coeffs.copy()
    quantized[np.abs(quantized) < threshold] = 0
    return quantized

# Apply global DCT
dct_transformed = dct(dct(image.T, norm='ortho').T, norm='ortho')  # 2D DCT
quantized_dct = quantize_dct(dct_transformed, threshold=50)  # Apply quantization
compressed_image_global = idct(idct(quantized_dct.T, norm='ortho').T, norm='ortho')  # 2D inverse DCT

# 2. Block-Based DCT Compression: Simulating Blocking Artifacts
def block_dct_compression(image, block_size=8, threshold=50):

    h, w = image.shape
    compressed_image = np.zeros((h, w), dtype=np.float32)
    for i in range(0, h, block_size):
        for j in range(0, w, block_size):
            block = image[i:i+block_size, j:j+block_size]
            dct_block = dct(dct(block.T, norm='ortho').T, norm='ortho')  # DCT
            quantized_block = np.where(np.abs(dct_block) < threshold, 0, dct_block)  # Quantization
            idct_block = idct(idct(quantized_block.T, norm='ortho').T, norm='ortho')  # Inverse DCT
            compressed_image[i:i+block_size, j:j+block_size] = idct_block
    return compressed_image

# Apply block-based DCT compression
compressed_image_block = block_dct_compression(image, block_size=8, threshold=50)

# 3. Visualize Results
plt.figure(figsize=(10, 6))

# Original Image
plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.axis('off')

# Global DCT Compression
plt.subplot(2, 3, 2)
plt.title("Global DCT Compression (Ringing)")
plt.imshow(np.clip(compressed_image_global, 0, 255), cmap="gray")
plt.axis('off')

# Artifacts from Global DCT Compression
plt.subplot(2, 3, 3)
plt.title("Artifacts (Ringing)")
plt.imshow(np.abs(image - compressed_image_global), cmap="hot")
plt.axis('off')

# Block-Based DCT Compression
plt.subplot(2, 3, 4)
plt.title("Block-Based DCT Compression (Blocking)")
plt.imshow(np.clip(compressed_image_block, 0, 255), cmap="gray")
plt.axis('off')

# Artifacts from Block-Based Compression
plt.subplot(2, 3, 5)
plt.title("Artifacts (Blocking)")
plt.imshow(np.abs(image - compressed_image_block), cmap="hot")
plt.axis('off')

# Difference Heatmap for Blocking Artifacts
plt.subplot(2, 3, 6)
plt.title("Artifacts (Zoomed Heatmap)")
plt.imshow(np.abs(image - compressed_image_block), cmap="hot")
plt.colorbar()
plt.axis('off')

plt.tight_layout()
plt.show()
```

**Common Artifacts in DCT Compression:**

Ringing Artifacts:

*   Appear as ripples or waves near sharp edges.

*   Caused by losing high-frequency details during compression.


Blocking Artifacts:


*   Appear as visible grid patterns at block boundaries.
*   Caused by processing the image in small, independent blocks (like 8×8 pixels) and mismatches after compression.




Both occur because DCT compression reduces file size by simplifying image details, which can distort the original content.
