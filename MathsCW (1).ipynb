```python
import numpy as np

def compute_partial_derivatives(x, y):
    """
    Computes the partial derivatives of L(x, y) at the given (x, y).
    Args:
        x (float): The x-coordinate.
        y (float): The y-coordinate.
    Returns:
        (float, float): Partial derivatives (∂L/∂x, ∂L/∂y).
    """
    partial_x = 2 * np.exp(2 * x) - y * np.sin(x * y)
    partial_y = 2 * y - x * np.sin(x * y)
    return partial_x, partial_y

# Example usage:
x_val, y_val = 1.0, 1.0
partial_x, partial_y = compute_partial_derivatives(x_val, y_val)
print(f"Partial derivatives at (x={x_val}, y={y_val}): ∂L/∂x = {partial_x}, ∂L/∂y = {partial_y}")
```

```python
def compute_gradient(x, y):
    """
    Computes the gradient vector of L(x, y) at the given (x, y).
    Args:
        x (float): The x-coordinate.
        y (float): The y-coordinate.
    Returns:
        (tuple): Gradient vector (∂L/∂x, ∂L/∂y).
    """
    return compute_partial_derivatives(x, y)

# Example usage:
gradient = compute_gradient(1.0, 1.0)
print("Gradient vector at (1,1):", gradient)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the loss function
def loss_function(x, y):
    return (np.exp(x))**2 + y**2 + np.cos(x * y)

# Generate grid points
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = loss_function(X, Y)

# Plot the surface
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('Loss Function L(x, y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('L(x, y)')
plt.show()
```

```python
def gradient_descent_demo():
    """
    Demonstrates gradient computation and its application in model optimization.
    """
    # Starting point
    x, y = 1.0, 1.0

    # Compute the gradient
    grad_x, grad_y = compute_gradient(x, y)

    print(f"Starting point: (x={x}, y={y})")
    print(f"Gradient vector: (∂L/∂x={grad_x}, ∂L/∂y={grad_y})")

    # Visualize gradient descent direction
    print("In gradient descent, we move opposite to this vector to minimize the loss.")

# Example usage:
gradient_descent_demo()
```

```python
import numpy as np

def series_ln_approximation(x, terms=10):
    """
    Approximate ln(1 + x) using the first 'terms' of the series.
    f(x) = x - x^2/2 + x^3/3 - x^4/4 + ...
    Args:
        x (float or np.ndarray): Input value(s) for x.
        terms (int): Number of terms in the series to use.
    Returns:
        float or np.ndarray: Approximated value of ln(1 + x).
    """
    approximation = np.zeros_like(x, dtype=float)
    for n in range(1, terms + 1):
        approximation += ((-1)**(n - 1)) * (x**n) / n
    return approximation

# Example usage:
x_vals = np.linspace(-1, 1, 100)
approx_vals = series_ln_approximation(x_vals, terms=10)
```

```python
import matplotlib.pyplot as plt

# Compute the actual values of ln(1 + x)
actual_vals = np.log(1 + x_vals)

# Plot the approximation and actual function
plt.figure(figsize=(10, 6))
plt.plot(x_vals, actual_vals, label="Actual ln(1 + x)", color="blue", linewidth=2)
plt.plot(x_vals, approx_vals, label="Approximation (10 terms)", color="orange", linestyle="--", linewidth=2)

# Add labels, legend, and title
plt.title("Series Approximation vs Actual Function", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
# Compute the absolute error between the approximation and the actual function
error = np.abs(actual_vals - approx_vals)

# Plot the error
plt.figure(figsize=(10, 6))
plt.plot(x_vals, error, label="Error", color="red", linewidth=2)

# Add labels, legend, and title
plt.title("Error Analysis of Series Approximation", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Error", fontsize=12)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.grid()

# Show the plot
plt.show()

# Print observations
print("Observations:")
print("- The series converges faster near x = 0.")
print("- As x approaches the boundaries of the interval [-1, 1], the error increases.")
print("- The series diverges outside the interval [-1, 1] due to the convergence properties of the logarithmic series.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
t = np.linspace(0, 1, 500, endpoint=False)  # Time vector
s = np.sin(2 * np.pi * t) + 0.5 * np.sin(10 * np.pi * t) + 0.3 * np.sin(50 * np.pi * t)  # Signal

# Perform Fourier Transform
freqs = np.fft.fftfreq(len(t), d=(t[1] - t[0]))  # Frequency axis
s_fft = np.fft.fft(s)  # Fourier Transform of the signal

# Plot the frequency spectrum
plt.figure(figsize=(10, 6))
plt.plot(np.abs(freqs[:len(freqs)//2]), np.abs(s_fft[:len(s_fft)//2]))
plt.title("Frequency Spectrum", fontsize=14)
plt.xlabel("Frequency (Hz)", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.grid()
plt.show()
```

```python
# Design a low-pass filter (frequencies above 20 Hz are set to 0)
cutoff = 20
s_fft_filtered = s_fft.copy()
s_fft_filtered[np.abs(freqs) > cutoff] = 0

# Perform Inverse Fourier Transform to get the filtered signal
s_filtered = np.fft.ifft(s_fft_filtered)

# Plot the original and filtered signals
plt.figure(figsize=(10, 6))
plt.plot(t, s, label="Original Signal", alpha=0.7)
plt.plot(t, s_filtered.real, label="Filtered Signal", linestyle="--", color="red")
plt.title("Original vs Filtered Signal", fontsize=14)
plt.xlabel("Time (s)", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("Image.png", cv2.IMREAD_GRAYSCALE)

# Perform 2D Fourier Transform
f_transform = np.fft.fft2(image)
f_shift = np.fft.fftshift(f_transform)  # Shift zero frequency to the center
magnitude_spectrum = 20 * np.log(np.abs(f_shift))  # Magnitude spectrum for visualization

# Create high-pass filter for edge detection
rows, cols = image.shape
crow, ccol = rows // 2, cols // 2  # Center point
mask = np.ones((rows, cols), np.uint8)
mask[crow-30:crow+30, ccol-30:ccol+30] = 0  # Remove low frequencies

# Apply the mask
f_shift_filtered = f_shift * mask
f_ishift = np.fft.ifftshift(f_shift_filtered)
image_edges = np.abs(np.fft.ifft2(f_ishift))

# Plot results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Magnitude Spectrum")
plt.imshow(magnitude_spectrum, cmap='gray')
plt.subplot(1, 2, 2)
plt.title("Edges Detected")
plt.imshow(image_edges, cmap='gray')
plt.show()
```

```python
# Create Gaussian Low-Pass Filter
def gaussian_low_pass(shape, sigma=10):
    rows, cols = shape
    x = np.linspace(-cols // 2, cols // 2, cols)
    y = np.linspace(-rows // 2, rows // 2, rows)
    X, Y = np.meshgrid(x, y)
    filter_gaussian = np.exp(-(X**2 + Y**2) / (2 * sigma**2))
    return filter_gaussian

# Apply Gaussian filter in frequency domain
gaussian_filter = gaussian_low_pass(image.shape)
f_shift_blur = f_shift * gaussian_filter
f_ishift_blur = np.fft.ifftshift(f_shift_blur)
image_blur_fft = np.abs(np.fft.ifft2(f_ishift_blur))

# Standard Gaussian blur using OpenCV
image_blur_cv = cv2.GaussianBlur(image, (15, 15), 10)

# Plot results
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title("Gaussian Blur (FFT)")
plt.imshow(image_blur_fft, cmap='gray')
plt.subplot(1, 3, 3)
plt.title("Gaussian Blur (OpenCV)")
plt.imshow(image_blur_cv, cmap='gray')
plt.show()
```

```python
import scipy.fftpack

# Apply DCT to the original image
dct = scipy.fftpack.dctn(image, type=2, norm='ortho')

# Resize image to 240x240
dct_resized = cv2.resize(dct, (240, 240), interpolation=cv2.INTER_AREA)

# Reconstruct the image using IDCT
idct_resized = scipy.fftpack.idctn(dct_resized, type=2, norm='ortho')

# Plot results
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title("Scaled DCT Image")
plt.imshow(idct_resized, cmap='gray')
plt.subplot(1, 3, 3)
plt.title("Difference (Original - DCT)")
plt.imshow(image - cv2.resize(idct_resized, image.shape[::-1]), cmap='gray')
plt.show()
```

```python
# Simulate compression by keeping only part of DCT coefficients
def compress_dct(image, keep_ratio=0.1):
    dct = scipy.fftpack.dctn(image, type=2, norm='ortho')
    rows, cols = dct.shape
    mask = np.zeros_like(dct)
    mask[:int(rows*keep_ratio), :int(cols*keep_ratio)] = 1  # Keep low frequencies only
    dct_compressed = dct * mask
    image_compressed = scipy.fftpack.idctn(dct_compressed, type=2, norm='ortho')
    return image_compressed

# Apply compression
compressed_image_10 = compress_dct(image, keep_ratio=0.1)
compressed_image_05 = compress_dct(image, keep_ratio=0.05)

# Plot results
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title("10% DCT Coefficients")
plt.imshow(compressed_image_10, cmap='gray')
plt.subplot(1, 3, 3)
plt.title("5% DCT Coefficients (Artifacts)")
plt.imshow(compressed_image_05, cmap='gray')
plt.show()
```
